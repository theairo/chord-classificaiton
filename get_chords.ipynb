```python
from pydub import AudioSegment
from pathlib import Path
from pedalboard import Pedalboard, Reverb
import soundfile as sf

directory = Path("Instruments")

notes_list = ["C", "C#", "D", "D#", "E", "F", "F#", "G", "G#", "A", "A#", "B"]

audio_count = 0

for audio_file in directory.rglob("*"):
    audio_count+=1

    audio = AudioSegment.from_file(audio_file)

    duration = len(audio)

    # Major triads
    for index in range(24):
        start_ms = index * 1000
        end_ms = (index+1)*1000
        chunk = audio[start_ms:end_ms]
        
        chunk.export(f"Chords\\{notes_list[index%12]}_{index}_{audio_count}.wav", format="wav")
        
        # Load audio
        audio2, sr = sf.read(f"Chords\\{notes_list[index%12]}_{index}_{audio_count}.wav")

        board = Pedalboard([
            Reverb(room_size=0.5, damping=0.5, wet_level=0.5)
        ])

        # Process audio
        wet_audio = board(audio2, sr)

        # Save output
        sf.write(f"Chords\\{notes_list[index%12]}_{index}_{audio_count}_rev.wav", wet_audio, sr)

    # Minor triads
    for index in range(24, 48):
        start_ms = index * 1000
        end_ms = (index+1)*1000
        chunk = audio[start_ms:end_ms]
        
        chunk.export(f"Chords\\{notes_list[index%12]}m_{index}_{audio_count}.wav", format="wav")

        # Load audio
        audio2, sr = sf.read(f"Chords\\{notes_list[index%12]}m_{index}_{audio_count}.wav")

        board = Pedalboard([
            Reverb(room_size=0.5, damping=0.5, wet_level=0.5)
        ])

        # Process audio
        wet_audio = board(audio2, sr)

        # Save output
        sf.write(f"Chords\\{notes_list[index%12]}m_{index}_{audio_count}_rev.wav", wet_audio, sr)
```
